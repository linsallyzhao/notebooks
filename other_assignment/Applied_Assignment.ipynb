```python
T = 1
n_steps = 252
delta_t = T / n_steps
S0 = 100
vol = 0.2
r = 0.05

path = np.random.randn(n_steps, 1)
path[0] = 0
path = (r - 0.5 * vol ** 2) * delta_t + vol * delta_t ** 0.5 * path
path = np.exp(path.cumsum()) * S0
running_avg = path.cumsum() / range(1, 253)
x = np.zeros(12)
y = np.zeros(12)
x[0] = 0
y[0] = path[0]
for i in range(3,252):

    if (i-1) % 22 == 0 :
        x[int((i-1)/22)] = i
        y[int((i-1)/22)] = path[i]
        
running_avg_dis = y.cumsum() / range(1, 13)

yy = np.zeros(252)
yy[: 22] = running_avg_dis[0]
for j in range(1, 12):
    yy[j*22 : (j+1) ** 22] = running_avg_dis[j]

plt.plot(path)
plt.plot(running_avg)
plt.scatter(x, y, c='r')
plt.scatter(x, running_avg_dis, c = 'g')
plt.plot(yy)
plt.xlabel("time", fontsize = 18)
plt.ylabel("Prices", fontsize = 18)
```
